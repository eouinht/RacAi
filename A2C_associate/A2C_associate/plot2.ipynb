```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        return False
    
plt.rcParams['text.usetex'] = True
plt.rc('font', family='sans-serif', size=33)
```

```python
candidate = ["Random", "RoundRobin", "NearestRU", "PPO", "MILP"]
figsize = (10, 10)
colors = ['grey','seagreen','teal', "cadetblue" ,'black']
markers = ['/','.','o','x', '']
linestyles = ["-", "--", "-.", ":", '-']
labels = ["Random-RU", "RoundRobin", "Nearest-RU", "PPO-GraphSAGE", "MILP"]
dpi = 100
bar_width = 20
bar_gap = 10
```

```python
# Total accept plot
item_name = "accept_hist"
plt.figure(figsize=figsize, dpi=dpi)
x = 0
x_ticks = []
plt.ylim(0,1)

plt.ylabel("Acceptance rate")
for i, c in enumerate(candidate):
    value = 0
    with open(f"results_final/run_20260119_221640/{c}/{item_name}_{c}.txt", "r") as f:
        lines = f.readlines()
        data = [list(map(float, line.strip().split())) for line in lines if is_number(line)]
        value = np.mean(data, axis=0)
        err = np.std(data, axis=0)
    value = float(value/20)
    plt.bar(
        x = x,
        height = value,
        # yerr = err,
        width = bar_width,
        color = 'w',
        edgecolor = colors[i],
        hatch = markers[i],
        label = labels[i],
        align = "center",
        zorder = 3
    )
    x_ticks.append(x)
    x += bar_width + bar_gap
plt.xticks(x_ticks, [])
# plt.xticks([])
plt.xlim([min(x_ticks)-bar_width, max(x_ticks)+bar_width])
plt.legend(loc = "upper left", framealpha=0)
plt.grid(color='lightgrey', linestyle='--', zorder=0)
plt.savefig(f"{item_name}.pdf", bbox_inches='tight')
plt.show()
```

```python
# Total reward
item_name = "reward_hist"
plt.figure(figsize=figsize, dpi=dpi)
x = 0
x_ticks = []
plt.ylim(0,300)

plt.ylabel("Average reward value")
for i, c in enumerate(candidate):
    value = 0
    with open(f"result_1/{c}/{item_name}_{c}.txt", "r") as f:
        lines = f.readlines()
        data = [list(map(float, line.strip().split())) for line in lines if is_number(line)]
        value = np.mean(data, axis=0)
    # value = float(value/30)
    plt.bar(
        x = x,
        height = value,
        width = bar_width,
        color = 'w',
        edgecolor = colors[i],
        hatch = markers[i],
        label = labels[i],
        align = "center",
        zorder = 3
    )
    x_ticks.append(x)
    x += bar_width + bar_gap
plt.xticks(x_ticks, [])
# plt.xticks([])
plt.xlim([min(x_ticks)-bar_width, max(x_ticks)+bar_width])
plt.legend(loc = "upper left", framealpha=0)
plt.grid(color='lightgrey', linestyle='--', zorder=0)
plt.savefig(f"{item_name}.pdf", bbox_inches='tight')
plt.show()
```

```python
# plot avg reward
plt.figure(figsize=(20,10), dpi=dpi)
with open(f"result_1/PPO/train/reward_hist_PPO.txt", "r") as f:
    lines = f.readlines()
    data = [float(line.strip()) for line in lines if is_number(line)]
    x_values = range(1, len(data)+1)
    avg_data = np.convolve(data, np.ones(20)/20, mode='valid')

for i in range(len(x_values)-len(avg_data)):
    avg_data = np.append(avg_data, avg_data[-1])

# đổi màu & độ đậm
plt.plot(x_values, data, color='gray', alpha=0.4, linewidth=1.0, linestyle='-', label='Reward')
plt.plot(x_values, avg_data, color='tab:orange', linewidth=4, linestyle='-', label='Average reward (window=$20$)')

plt.xlabel('Episode')
plt.ylabel('Reward')
plt.legend(loc = "lower right", framealpha=0)
plt.grid(color='lightgrey', linestyle='--', zorder=0)
plt.ylim(0,175)
plt.yticks(np.arange(0, 200, 25))
plt.xlim(0,5000)
plt.savefig(f"reward_hist_PPO.pdf", bbox_inches='tight')
plt.show()
```
